#what is first-order and 2nd order differencing?
2nd order differencing might have qudratic trend. 

```python
from statsmodels.tsa.statespace.tools import diff
```

```python
#1st Order differencing
diff(df['b'], k_diff=1).plot()
```

```python
#2nd Order differencing
diff(df['b'], k_diff=2).plot()
```

```python
#10th Order differencing
diff(df['b'], k_diff=10).plot()
```

```python
#20th Order differencing
diff(df['b'], k_diff=20).plot()
```
